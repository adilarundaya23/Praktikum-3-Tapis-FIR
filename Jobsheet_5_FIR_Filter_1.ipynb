Gambar tersebut adalah grafik yang membandingkan dua jenis filter lowpass: Kaiser dan Equal Ripple (Remez), berdasarkan respons frekuensi dalam desibel (dB) di sepanjang rentang frekuensi dalam kilohertz (kHz). Pada sumbu x, frekuensi berkisar dari 0 hingga 0,5 kHz, dan pada sumbu y, gain filter dipetakan dari 0 hingga -70 dB.

Garis biru mewakili filter Equal Ripple (Remez), dengan 53 taps, ditandai dengan pola respons yang bergetar, yang berulang kali naik turun sambil secara umum menurun di sepanjang spektrum frekuensi.
Grafik tersebut menyoroti perbedaan kinerja dalam bagaimana setiap jenis filter meredam frekuensi, dengan perhatian khusus pada kecepatan dan pola redaman di sepanjang spektrum.

```python

# Desain filter FIR high-pass menggunakan metode Kaiser window
# Parameter:
# 1/8: Frekuensi cutoff (normalized cutoff frequency)
# 1/6: Frekuensi transisi (normalized transition band width)
# 50: Panjang filter (number of taps)
# 1.0: Faktor beta untuk jendela Kaiser (beta parameter for the Kaiser window)
b_k_hp = fir_d.firwin_kaiser_hpf(1/8, 1/6, 50, 1.0)

# Desain filter FIR high-pass menggunakan metode Remez (Parks-McClellan)
# Parameter:
# 1/8: Frekuensi cutoff (normalized cutoff frequency)
# 1/6: Frekuensi transisi (normalized transition band width)
# 0.2: Ripple toleransi (passband ripple) dalam dB
# 50: Panjang filter (number of taps)
# 1.0: Berat toleransi ripple (weighting of the ripple)
b_r_hp = fir_d.fir_remez_hpf(1/8, 1/6, 0.2, 50, 1.0)
```

```python
# Menghitung dan menggambarkan respon frekuensi dari dua filter, yaitu Kaiser dan Equal Ripple
fir_d.freqz_resp_list([b_k_hp, b_r_hp], [[1], [1]], 'dB', fs=1)

# Mengatur batas vertikal pada plot dari -80 dB hingga 5 dB
plt.ylim([-80, 5])

# Menambahkan judul pada plot
plt.title(r'Kaiser vs Equal Ripple Lowpass')

# Menambahkan label pada sumbu y untuk menunjukkan Gain Filter dalam dB
plt.ylabel(r'Filter Gain (dB)')

# Menambahkan label pada sumbu x untuk menunjukkan Frekuensi dalam kHz
plt.xlabel(r'Frequency in kHz')

# Menambahkan legenda untuk membedakan antara filter Kaiser dan Equal Ripple
plt.legend((r'Kaiser: %d taps' % len(b_k), r'Remez: %d taps' % len(b_r)), loc='best')

# Menambahkan grid pada plot untuk memudahkan pembacaan
plt.grid()
```

```python
# Desain filter FIR dengan jendela Kaiser menggunakan fungsi firwin_kaiser_bpf
# Parameter:
# 7000, 8000: frekuensi cut-off bagian bawah (mulai dan akhir transisi) dalam Hz
# 14000, 15000: frekuensi cut-off bagian atas (mulai dan akhir transisi) dalam Hz
# 50: jumlah koefisien filter (order filter + 1)
# 48000: frekuensi sampling dalam Hz
b_k_bp = fir_d.firwin_kaiser_bpf(7000, 8000, 14000, 15000, 50, 48000)

# Desain filter FIR dengan metode Remez menggunakan fungsi fir_remez_bpf
# Parameter:
# 7000, 8000: frekuensi cut-off bagian bawah (mulai dan akhir transisi) dalam Hz
# 14000, 15000: frekuensi cut-off bagian atas (mulai dan akhir transisi) dalam Hz
# 0.2: ripple dalam band stop (deviasi maksimal dalam passband dan stopband)
# 50: jumlah koefisien filter (order filter + 1)
# 48000: frekuensi sampling dalam Hz
b_r_bp = fir_d.fir_remez_bpf(7000, 8000, 14000, 15000, 0.2, 50, 48000)
```

**Traditional IIR Filter Design using the Bilinear Transform**

```python
# Parameter filter
fs = 48000  # Frekuensi sampling
f_pass = 5000  # Frekuensi batas bawah jalur lintasan (passband)
f_stop = 8000  # Frekuensi batas atas jalur hentian (stopband)

# Merancang filter Butterworth
b_but, a_but, sos_but = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'butter')

# Merancang filter Chebyshev tipe 1
b_cheb1, a_cheb1, sos_cheb1 = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'cheby1')

# Merancang filter Chebyshev tipe 2
b_cheb2, a_cheb2, sos_cheb2 = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'cheby2')

# Merancang filter elliptic (Cauer)
b_elli, a_elli, sos_elli = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'ellip')
```

```python
# Memanggil fungsi freqz_resp_cas_list dari modul iir_d
# Fungsi ini mengambil daftar filter IIR dalam bentuk Second-Order Sections (SOS) dan
# memplot respons frekuensi dari masing-masing filter.
# fs=48 adalah frekuensi sampel yang digunakan dalam filter.
iir_d.freqz_resp_cas_list([sos_but, sos_cheb1, sos_cheb2, sos_elli], 'dB', fs=48)

# Menentukan batas-batas sumbu y pada plot
plt.ylim([-80, 5])

# Memberi judul pada plot
plt.title(r'IIR Lowpass Compare')

# Memberi label pada sumbu y
plt.ylabel(r'Filter Gain (dB)')

# Memberi label pada sumbu x
plt.xlabel(r'Frequency in kHz')

# Menambahkan legenda pada plot yang menjelaskan masing-masing filter
plt.legend((r'Butter order: %d' % (len(a_but) - 1),  # Menampilkan orde filter Butterworth
            r'Cheby1 order: %d' % (len(a_cheb1) - 1),  # Menampilkan orde filter Chebyshev Tipe 1
            r'Cheby2 order: %d' % (len(a_cheb2) - 1),  # Menampilkan orde filter Chebyshev Tipe 2
            r'Elliptic order: %d' % (len(a_elli) - 1)),  # Menampilkan orde filter Elip
           loc='best')  # Menempatkan legenda pada posisi terbaik

# Menambahkan grid pada plot
plt.grid()
```

Gambar di atas adalah grafik perbandingan respons filter lowpass IIR dengan berbagai jenis dan urutan. Sumbu horizontal mewakili frekuensi dalam kHz, sedangkan sumbu vertikal mewakili keuntungan filter (Filter Gain) dalam dB.

Grafik ini membandingkan empat jenis filter lowpass IIR dengan urutan yang berbeda:

Garisan oranye: Filter Chebyshev 1 dengan urutan 8.

**Amplitude Response Bandpass Design**

```python
# Memanggil fungsi fir_remez_bpf dari modul fir_d untuk mendesain filter bandpass dengan metode Remez
# Parameter yang diberikan adalah batas bawah dan atas dari bandpass (23 kHz - 24 kHz dan 28 kHz - 29 kHz), ripple maksimum (0.5 dB),
# redaman minimum (70 dB), frekuensi sampling (96 kHz), dan panjang filter (8)
b_rec_bpf1 = fir_d.fir_remez_bpf(23000,24000,28000,29000,0.5,70,96000,8)

# Memanggil fungsi freqz_resp_list dari modul fir_d untuk mendapatkan respons frekuensi dari filter yang telah dibangun
# Parameter yang diberikan adalah koefisien filter FIR yang telah dibuat, [1] untuk menunjukkan filter tersebut adalah filter FIR
# dan mode='dB' untuk mendapatkan respons dalam skala desibel, fs=96000 adalah frekuensi sampling, dan plt.ylim untuk membatasi rentang sumbu y.
fir_d.freqz_resp_list([b_rec_bpf1],[1],mode='dB',fs=96000)
plt.ylim([-80, 5])

# Menambahkan grid pada plot
plt.grid();
```

Gambar ini adalah grafik yang menunjukkan respons frekuensi dalam domain magnitudo. Sumbu horizontal (x) mewakili frekuensi dalam satuan Hertz (Hz), sementara sumbu vertikal (y) mewakili gain dalam satuan desibel (dB).


Grafik menunjukkan adanya gain konstan sekitar 0 dB untuk rentang frekuensi sekitar 20,000 Hz hingga 30,000 Hz.
Di luar rentang tersebut (baik di bawah 20,000 Hz maupun di atas 30,000 Hz), gain turun drastis menjadi sekitar -80 dB.


Grafik memiliki bentuk seperti filter pita, yang memungkinkan frekuensi tertentu (20,000 Hz hingga 30,000 Hz) lewat dengan gain tinggi, sementara memblokir frekuensi di luar rentang tersebut dengan gain rendah.

```python
# Merancang filter FIR menggunakan metode Remez
# Parameter pertama hingga keempat menunjukkan batas frekuensi passband (dalam Hz)
# Parameter kelima adalah lebar band reject (dalam Hz)
# Parameter keenam adalah redaman di passband (dB)
# Parameter ketujuh adalah frekuensi sampel (dalam Hz)
# Parameter kedelapan adalah jumlah koefisien filter
b_rec_bpf1 = fir_d.fir_remez_bpf(23000, 24000, 28000, 29000, 0.5, 70, 96000, 8)

# Menghitung respons filter FIR terhadap frekuensi
# Parameter pertama adalah koefisien filter FIR
# Parameter kedua adalah 1 karena ini adalah filter FIR, bukan IIR
# Parameter ketiga adalah mode 'groupdelay_s', yang menunjukkan plot respons group delay
# Parameter keempat adalah frekuensi sampel (fs) dalam Hz
fir_d.freqz_resp_list([b_rec_bpf1], [1], mode='groupdelay_s', fs=96000)

# Menambahkan grid pada plot
plt.grid()

# Menampilkan plot
plt.show()
```

Sumbu vertikal menunjukkan delay kelompok (group delay) dalam satuan sampel, yang berkisar dari 0 hingga 140 sampel.

Kurva pada grafik ini berupa garis horizontal dengan nilai konstan sekitar 120 sampel pada semua frekuensi yang diukur, menunjukkan bahwa delay kelompok tidak berubah dengan frekuensi. Hal ini bisa mengindikasikan bahwa sistem memiliki delay konstan untuk semua frekuensi yang diuji.

```python
# Memanggil fungsi untuk membuat filter bandpass dengan parameter yang diberikan
# Frekuensi cutoff ditentukan oleh rentang 23000-24000 Hz dan 28000-29000 Hz
# 0.5 adalah gain, 70 adalah ordernya, dan 96000 adalah frekuensi samplingnya
# 'ellip' mengindikasikan penggunaan metode eliptik untuk desain filter
b_rec_bpf2, a_rec_bpf2, sos_rec_bpf2 = iir_d.IIR_bpf(23000, 24000, 28000, 29000, 0.5, 70, 96000, 'ellip')

# Menggunakan 'np.errstate' untuk mengabaikan pesan kesalahan saat pembagian nol terjadi
with np.errstate(divide='ignore'):
    # Menggunakan filter yang telah dibuat untuk mendapatkan respons frekuensi
    # Memanggil fungsi 'freqz_resp_cas_list' dari 'iir_d' dengan parameter 'sos_rec_bpf2'
    # Mode 'dB' menentukan keluaran berupa respons dalam bentuk desibel
    # fs=96000 adalah frekuensi sampling
    iir_d.freqz_resp_cas_list([sos_rec_bpf2], mode='dB', fs=96000)

# Menetapkan batas sumbu y dari plot
plt.ylim([-80, 5])

# Menambahkan grid pada plot
plt.grid()
```
